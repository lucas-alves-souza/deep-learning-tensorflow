```python
import tensorflow as tf
```

```python
a = tf.add(2, 2, name = 'add')
b = tf.multiply(a, 3, name = 'mult1')
c = tf.multiply(b, a, name = 'mult2')
```

Em modo *eager*, os grafos são construídos e destruídos em tempo real.

```python
c.graph
```

Existe um grafo padrão, mas ele não é utilizado.

```python
tf.compat.v1.get_default_graph()
```

Verificando o grafo construído em tempo real.

```python
writer = tf.summary.create_file_writer('/content/output')
a = tf.add(2, 2, name = 'add')
b = tf.multiply(a, 3, name = 'mult1')
c = tf.multiply(b, a, name = 'mult2')
writer.close()
```

```python
with tf.name_scope('Operacoes'):
    writer = tf.summary.create_file_writer('/content/output')
    with tf.name_scope('Escopo_A'):
        a = tf.add(2, 2, name = 'add')
    with tf.name_scope('Escopo_B'):
        b = tf.multiply(a, 3, name = 'mult1')
        c = tf.multiply(b, a, name = 'mult2')
    writer.close()
```

É possível forçar a execução dentro de um grafo explícito (em modo *lazy*):

```python
grafo = tf.Graph()
```

```python
grafo
```

```python
with grafo.as_default():
    a = tf.add(2, 2, name = 'add')
    b = tf.multiply(a, 3, name = 'mult1')
    c = tf.multiply(b, a, name = 'mult2')
    print(c.graph is grafo)
```
